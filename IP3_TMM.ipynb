```python
df2['Date'] = [d.strftime('%m-%d-%Y') if not pd.isnull(d) else '' for d in df2['Date']]
df3['Date'] = [d.strftime('%m-%d-%Y') if not pd.isnull(d) else '' for d in df3['Date']]
df4['Date'] = [d.strftime('%m-%d-%Y') if not pd.isnull(d) else '' for d in df4['Date']]
```

```python
df1.insert(0,"Company",'50Hertz')
```

```python
df1.head()
```

```python
df2.insert(0,"Company",'Amprion')
df3.insert(0,"Company",'TenneTTSO')
df4.insert(0,"Company",'TransnetBW')
```

```python
df=[df1,df2,df3,df4]
```

```python
df = pd.concat(df)
```

```python
df.sample(10)
```

```python
df.count()
```

```python
df.Date=pd.to_datetime(df.Date)
```

```python
df.dtypes
```

```python
year = df['Date'].dt.year
```

```python
company_year = pd.DataFrame(df.groupby(['Company',year]).mean())
```

```python
company_year
```

```python
noon = pd.DataFrame(company_year['12:00:00'])
```

```python
noon
```

```python
midnight = pd.DataFrame(company_year['00:00:00'])
```

```python
midnight
```

```python
noon_bar = noon.plot(kind='barh'),plt.title("TW/Hours by Company/Year at Noon")
```

```python
midnight_bar = midnight.plot(kind='barh'),plt.title("TW/Hours by Company/Year at Midnight")
```

```python
fig, axarr = plt.subplots(2, figsize=(6,4), sharex=True)
noon, midnight = noon.align(midnight, axis=1)
noon.plot.barh(ax=axarr[0], title='TW/Hours by Company/Year at noon', ylabel=None, legend=None)
midnight.plot.barh(ax=axarr[1], title='TW/Hours by Company/Year at midnight', legend=None)
```
